````python
import random as rnd
import math
import matplotlib.pyplot as plt
point = [(rnd.uniform (-1,1), rnd.uniform(-1,1)) for _ in range(5000)]

point_red =[]
point_blue = []

for n in range (5000):
    if math.dist([0,0],list(point[n])) <=1:
        point_red.append(point[n])
    else:
        point_blue.append(point[n])


x, y = zip(*point_red)
a, b = zip(*point_blue)

plt.figure(figsize=(6,6))
plt.scatter (x, y,c="r")  
plt.scatter (a,b)

fraction = len(point_red)/5000
print(fraction)
Converage = math.pi/4
print (Converage)
````

## 7. A cute rabbit among two ferocious snakes (**)

There are three identical doors, behind two of them, there are ferocious snakes and behind one of them is a cute little rabbit. Your aim is to pick a door and get the cute rabbit. You start picking one door randomly and suddenly another door opens and snake slithers away *pssss*. Now you have a choice to either stay with your door or change to the remaining door. 

<img src="../assets/snakes_rabbits.png" width = 200>

&nbsp; a) What is your choice and why? (**)

&nbsp; b) Randomly assign which door has the rabbit, and randomly pick an initial door. Now simulate 10, 100, 1000, 10000, 100000, 1000000 times for staying and for switching doors. Calculate the proportion of finding the rabbits for each set of simulations. Plot the results (**)


<details>

<summary>Hint</summary>

Try to reason using probability.


</details>

````python
# change to the remaining door. because om you do not change proportion of finding the rabbits is 1/3. if you change proportion of finding the rabbits is 2/3

import random
import matplotlib.pyplot as plt
import math



i = [10**n for n in range (1,7)]
simulate_stay= []
simulate_change= []

for m in i:
    win_stay, win_change = 0,0
    for _ in range (m):

        door_3 = ["rabbit", "snake", "snake"]
        random.shuffle (door_3)
        first_pick = random.choice(door_3)

        if first_pick == "rabbit":
            door_2 = [first_pick,"snake"]
        else:
            door_2 = [first_pick,"rabbit"]

        change_pick = door_2[1]

    
        if first_pick == "rabbit":
            win_stay +=1
        if change_pick =="rabbit":
            win_change +=1

    P_stay, P_change = (win_stay/m), (win_change/m)
    simulate_stay.append(P_stay)
    simulate_change.append(P_change)
print (simulate_change,simulate_stay)
    
plt.plot (simulate_stay,color = "red")
plt.plot (simulate_change, color = "blue")
plt.xticks([0,1,2,3,4,5],i)
plt.xlabel("simulation")
plt.ylabel("propertion to find rabbit")
plt.title('propertion to find rabbit for different simulation')




````

````python
import random

i = [10**n for n in range(1,7)]

def trial():
    door_3 = ["rabbit", "snake", "snake"]
    random.shuffle(door_3)
    first_pick = random.choice(door_3)
    # 如果先选到 rabbit，则换到 snake；否则换到 rabbit
    change_pick = "snake" if first_pick == "rabbit" else "rabbit"
    return first_pick == "rabbit", change_pick == "rabbit"

def simulate_probs(m):
    outcomes = [trial() for _ in range(m)]        # 生成 m 次试验（每次为 (stay_win, change_win)）
    stay_prob = sum(1 for s, _ in outcomes if s) / m
    change_prob = sum(1 for _, c in outcomes if c) / m
    return stay_prob, change_prob

simulate = [simulate_probs(m) for m in i]
simulate_stay  = [s for s, _ in simulate]
simulate_change = [c for _, c in simulate]

print("simulate_stay:", simulate_stay)
print("simulate_change:", simulate_change)
````
