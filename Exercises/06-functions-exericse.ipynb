## 1. Area (*)

Create a function that takes the base and height of a triangle as input parameters and returns the area of the triangle.

````python
def Area_triangle(b:float,h:float):

    return b*h/2


a = input ("Please enter base and height of a triangle(cm), use black between two numbers:")
b,h = [float(i) for i in a.split()]

print (f"The area of triangle with base and height ({b}cm and {h}cm) is {Area_triangle(b,h)} cm2")
````

## 2. Euclidean distance (*)

The formula for Euclidean distance in 2D between $P: (p_1, p_2)$ and $Q: (q_1, q_2)$  is: 

$d(P,Q) = \sqrt{(p_1-q_1)^2 + (p_2-q_2)^2}$

&nbsp; a) &nbsp; Create a function that takes two points as input parameters and return the Euclidean between them. (*)

&nbsp; b) &nbsp; Let the user input two points. Call the function using the users input points. (*)

&nbsp; c) &nbsp; Use your function to calculate distances between the origin (0, 0) and each of these points: (10, 3), (-1, -9), (10, -10), (4, -2), (9, -10). (*)

<details>

<summary>Hint</summary>

Use a **for** statement

</details>
<br>
<details>

<summary>Answer</summary>

c)

```
[10.4, 9.1, 14.1, 4.5, 13.5]
```

</details>


````python
import numpy as np
def enclidean (a,b):
    return np.sqrt (np.sum((a-b)**2))

c = input (" Please enter two point, use ',' between x, y and blank between point")
a,b = c.split()
a = np.array ([float(i) for i in a.split(",")])
b = np.array ([float(i) for i in b.split(",")])
print (f"Distance between {a} and {b} is {enclidean(a,b):.2f}")

P = np.array([0,0])
Q = np.array([[10,3],[-1,-9],[-10,-10], [4,-2], [9,-10]])
D =[enclidean(P,q) for q in Q]
print (tuple(round(x,2) for x in map(float,D)))
````

- 1st 50-lapp
- 1st 20-lapp
- 1st 10-krona

- 2st 2-krona

Now let the user input a value, and use the function to calculate the change.
